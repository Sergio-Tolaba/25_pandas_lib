# Analisis exploratorio de datos - EDA
## ** Demanda de ML (Machine Learning)**

¿Cuáles son los valores promedio de alquiler por tipo de propiedad?
* Valores promedio de alquiler por tipo de inmueble

```python
datos.head()
```

```python
# El promedio general de alquileres se extrael de la columna Valor
datos['Valor'].mean()
```

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_tipo_prom = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_prom.plot(kind= 'barh',figsize=(12,8), color='red')
```

* Porcentaje de cada tipo de inmueble presente en la base de datos

```python
# En cada tipo de inmueble, cada tipo se repite ej Cocineta, Cocineta, etc
datos.Tipo.unique()
```

```python
inmueble_residencial = ['Cocineta','Posada/Chalé', 'Casa de Vecindad','Estudio','Loft','Casa de Condominio','Rancho','Casa','Departamento','Departamento en Hotel']
```

```python
df = datos.query('@inmueble_residencial in Tipo') # datos.query('@inmueble_residencial not in Tipo')
df.head()
```

```python
df_tipo_prom = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_prom.plot(kind= 'barh',figsize=(12,8), color='red')
```

* Porcentaje de cada tipo de inmueble presente en la base de datos

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts()
```

```python
porcentaje_tipo =df.Tipo.value_counts(normalize=True).to_frame() # Sólo modelaremos con Departamentos porque representan el 85% de los inmuebles residenciales (ML)
porcentaje_tipo.plot(kind= 'bar',figsize=(12,8), color='green', xlabel='Tipo de inmueble', ylabel='Porcentaje')
```
